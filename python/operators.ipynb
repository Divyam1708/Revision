<h2>Aritmetic</h2>

```python
#Declaration for using in Arithmetic operators
int_x=7
int_y=4
float_x=4.657
float_y=6.875
complex_x=(8+1j)
complex_y=6-4j
bool_x=True
bool_y=False
```

```python
print(f'int+float=float')
print(int_x+float_x)
```

```python
print(f'int+complex=complex')
print(int_x+complex_x)
```

```python
print(f'float+complex=complex')
print(float_x+complex_y)
```

```python
print(f'float+bool=float')
print(float_x+bool_x)
```

```python
print(f'bool+complex=complex')
print(complex_x+bool_x)
```
