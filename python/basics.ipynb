<h3>Typecasting</h3>

<h4>int => float,str,bool</h4>

```python
int_y=7
print('float(int_y): ',float(int_y))
print('str(int_y): ',str(int_y))
print('bool(int_y): ',bool(int_y))
```

<h4>float => int,str,bool</h4>

```python
float_y=7.9876
print('int(float_y): ',int(float_y))
print('str(float_y): ',str(float_y))
print('bool(float_y): ',bool(float_y))
```

<h4>str => int,float,bool</h4>

<h5>1. str declared with integer numeric value</h5>

```python
str_y='798'
print('int(str_y): ',int(str_y))
print('float(str_y): ',float(str_y))
print('bool(str_y): ',bool(str_y))
```

<h5>2. str declared with non-numeric values</h5>

```python
str_a_with_truly_nonNumeric_value='Hello'
print('bool(str_a_with_truly_nonNumeric_value): ',bool(str_a_with_truly_nonNumeric_value))
print('int(str_a_with_truly_nonNumeric_value): ',int(str_a_with_truly_nonNumeric_value))
print('float(str_a_with_truly_nonNumeric_value): ',float(str_a_with_truly_nonNumeric_value))

```

<h5>3. str declared with numeric but non-integer values</h5>

```python
str_a_with_numeric_but_nonInteger_value='98.987'
print('float(str_a_with_numeric_but_nonInteger_value): ',float(str_a_with_numeric_but_nonInteger_value))
print('bool(str_a_with_numeric_but_nonInteger_value): ',bool(str_a_with_numeric_but_nonInteger_value))
print('int(str_a_with_numeric_but_nonInteger_value): ',int(str_a_with_numeric_but_nonInteger_value))
```

<h4>bool => int,float,str</h4>

```python
bool_x_true=True
print('int(bool_x_true): ',int(bool_x_true))
print('float(bool_x_true): ',float(bool_x_true))
print('str(bool_x_true): ',str(bool_x_true))
bool_y_false=False
print('int(bool_y_false): ',int(bool_y_false))
print('float(bool_y_false): ',float(bool_y_false))
print('str(bool_y_false): ',str(bool_y_false))
```
